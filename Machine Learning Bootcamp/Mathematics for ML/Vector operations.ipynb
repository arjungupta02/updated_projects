```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, color = 'b', scale_units = 'xy',angles = 'xy',scale = 1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, color = 'b', scale_units = 'xy',angles = 'xy',scale = 1)
plt.quiver(0,0,-3,6, color = 'g', scale_units = 'xy',angles = 'xy',scale = 1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition

```python
vec1 = np.array([0,0,4,5])
vec2 = np.array([0,0,-3,6])

vec1 + vec2
```

```python
plt.quiver(0,0,4,5, color = 'b', scale_units = 'xy',angles = 'xy',scale = 1)
plt.quiver(0,0,-3,6, color = 'g', scale_units = 'xy',angles = 'xy',scale = 1)
plt.quiver(0,0,-1,11, color = 'r', scale_units = 'xy',angles = 'xy',scale = 1)
plt.xlim(-8,13)
plt.ylim(-8,13)
plt.show()
```

Multiplication of vector by scalar

```python
vec1 = np.array([0,0,4,5])
vec2 = 2*vec1

vec2
```

```python
plt.quiver(0,0,4,5, color = 'g', scale_units = 'xy',angles = 'xy',scale = 1)
plt.quiver(0,0,8,10, color = 'r', scale_units = 'xy',angles = 'xy',scale = 1)
plt.xlim(-8,13)
plt.ylim(-8,13)
plt.show()
```

Dot product (a*b) + (c*d)

```python
a = np.array([4,5])
b = np.array([-3,7])

a_dot_b = np.dot(a,b)
a_dot_b
```

Cross product

```python
a = np.array([7,4,5])
b = np.array([10,-3,7])

a_cross_b = np.cross(a,b)
a_cross_b
```
