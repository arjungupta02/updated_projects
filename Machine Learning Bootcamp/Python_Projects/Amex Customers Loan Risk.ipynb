### Understanding the Loan Default System

**How does a Loan Credit system works?**

* Loans are paid in pre-defined increments over the term defined. Say you make monthly payments towards your car loan, each payment will cover the interest due and some amount of the principal. 
* The more money you can apply to a payment means more principal you knock out in each payment. 
* Paying down your principal and wrapping up a loan quickly means you can save money you would have spent on interest payments.

Missing the minimum payment due date leads to a late payment, often accompanied by late fees. In addition, continued delay might lead to default. 

**What is defaulter?**

A person is considered a defaulter by the bank when they fail to make the required payments on their credit card or loan as per the agreed-upon terms. In the context of Loans, it can be of two types : Secured which are secured by a collateral (house or vehicle) and unsecured (Personal loans not secured by anything). The **personal loan** will be classified as a non-performing asset if the borrower defaults beyond 90 days. If the **loan is not repaid even after 180 days**, the lender has the right to proceed legally against the borrower under Section 138 of the Negotiable Instruments Act 1881. Before proceeding legally, the lender will exercise all recovery measures like sending two reminders followed by a recall letter. The recall letter will be the final reminder wherein it will be specifically mentioned that the lender will proceed legally if the dues are not cleared within a specific date. If you do not repay your loan, **the lender can take you to court.** The court will then require that you pay back the amount in full or face other penalties such as wage garnishment or seizure of assets. The lender may also report the debt to credit bureaus and send debt collectors after you if payments become overdue.

**Identifying Missing values**

```python
features_with_na = [feature for feature in df.columns if df[feature].isnull().sum()> 0]

for i in features_with_na:
    print(i, 'has', np.round(df[i].isnull().mean(),4), '% missing values')
    #some %s are very less in exponent and hence it's almost 0.0 when rounded off 
```

```python
for feature in features_with_na:
    
    data = df.copy()
    data[feature] = np.where(data[feature].isnull(),1,0)
    data.groupby(by = feature)['repay_fail'].count().plot.bar()  #count of nulls for both failed and completed payments 
    plt.title(feature)
    plt.show()
```

Here with the relation b/w the missing values and dependant variable "repay_fail" is clearly visible. There are very less null values (sometimes even 1) in loan_amnt, funded_amnt but in some columns like "next_pymnt_d" it is quite huge (35097 values) We need to replace these nan value with something meanginful as discussed in Feature Engineering section.

We can see almost all our continous features are **Right skewed** so we will havw to either convert them in Normal distribution or keep it as it is while dealing with outliers 

```python
#since some features do not follow gaussian distribution we can perform log transformation 

import scipy.stats as stat
import pylab

for feature in continuous_features:
    data = df.copy()
    data[feature+'_exp'] = data[feature] ** (1/2) #square root transformation to normal distribution
    sns.distplot(data[feature+'_exp'], hist = True, kde = False)
    plt.title(feature)
    plt.show()
        
# a = 3
# if a>5:
#     pass
# else 
#     print('a is less than 10')
    
#     #pass statement is used as a null statement
#     #so if a condition is true nothing happens 

#np.log(df['loan_amnt'].unique())  #log transformation of 0 would be infinity so we can't visually plot this
#only total_acc doesn't have 0 so it could be successfully converted into log transformation
```
