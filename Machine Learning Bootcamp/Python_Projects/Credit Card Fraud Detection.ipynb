Build a sample dataset with similar distribution of Normal and Fraud transactions

Number of Fraudulent Transactions --> 492

```python
legit_sample = legit.sample(n = 492)
```

```python
legit_sample['Amount'].describe()
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head(5)
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
new_dataset.groupby('Class').mean()['Amount']
```

Splitting the data into Features and Targets

```python
#Features - Input to determine the target : Everything except Class -Independent
#Targets - what we are determining : Class - Dependent
```

```python
#splitting into features and targets
X = new_dataset.drop('Class', axis = 1)
Y = new_dataset['Class']
```

```python
X.head()
```

```python
Y.head()
```

```python
#splitting into training and testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
#stratify - equal dist of y in both train and test 
#random_state = 2 (classification)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

Model Training

Logisitic Regression

```python
model = LogisticRegression()
```

```python
#training the logistic regression data with training data 
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
training_data_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(training_data_prediction, Y_train)
```

```python
print('The accuracy of training data is :', training_data_accuracy)
```

Evaulating testing data

```python
testing_data_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(testing_data_prediction, Y_test)
```

```python
print('The accuracy of testing data is :', testing_data_accuracy)
```
