Linear Regression :

Y = wx + b (Y= mx + b)
 
Y --> Dependent Variable
X --> Independent Variable
w --> Weight (Slope)
x --> Bias (Intercept) 

Gradient Descent :

Gradient Descent is an optimisation algorithm used for minimising the loss func in various ML algorithms. It is used for updating the parameters (m,b)

w = w - l(dw)


b = b - l(db)


Learning rate (l):

Lerning rate is a tuning parameter in an optimisation algorthm that determines the step size at each iteration while moving toward a min loss func
