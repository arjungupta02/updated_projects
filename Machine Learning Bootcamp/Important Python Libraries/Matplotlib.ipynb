Matplotlib : useful for making plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z, 'r.')
plt.xlabel('angle')
plt.ylabel('cos value')
plt.title('cos wave')
plt.show()
```

```python
x = np.linspace(-10,10,100)
y = x**2
```

```python
#parabola
plt.plot(x,y,'g+')
plt.show()
```

```python
#multiple line charts
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'r')
plt.plot(x, np.cos(x), 'g.')
plt.show()
```

Bar chart

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
languages = ['Spanish', 'English', 'Hindi', 'Chinese','Telgu']
count = [24,80,67,29,90]
axes.bar(languages, count)
plt.xlabel('Languages')
plt.ylabel('Number of people')
plt.show()
```

Pie Chart

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
languages = ['Spanish', 'English', 'Hindi', 'Chinese','Telgu']
count = [24,80,67,29,90]
axes.pie(count, labels = languages, autopct = '%1.1f%%')
plt.show()
```

Scatter plot

```python
x = np.linspace(1,10,30)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.scatter(x,y,color ='r')
axes.scatter(x,z, color = 'g')
plt.show()
```

```python
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)

fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.scatter(x,y,z,cmap= 'blue')
plt.show()
```
